```python
import numpy as np
import random
import gym


environment = gym.make("Taxi-v2")
environment.render()

state_size = environment.observation_space.n
action_size = environment.action_space.n

qtable = np.zeros((state_size, action_size))

total_episodes = 50000
max_steps = 99

learning_rate = 0.7
gamma = 0.618

epsilon = 1.0
max_epsilon = 1.0
min_epsilon = 0.01
decay_rate = 0.01


def qlearn(environment, state_size, action_size, qtable, total_episodes, max_steps, learning_rate, gamma, epsilon,
                                                             max_epsilon, min_epsilon, decay_rate):
        """
        A generic Q-learning algorithm for the reinforcement-learning of the agent.
        """
        # Rewards variable to count the total rewards
        rewards = []

        for episode in range(total_episodes):
            # Reset the environment
            state = environment.reset()
            # Reset the step
            step = 0
            # Agent is not done yet (done means dead, completed the goal etc. depending on a game/task)
            done = False
            # Total rewards which is the sum of all the rewards per episode
            total_rewards = 0
            
            # For each step per episode (number of steps is 'max_steps')
            for step in range(max_steps):
                # Exploration-exploitation tradeoff
                tradeoff = random.uniform(0, 1)
                
                # Do the exploitation
                if tradeoff > epsilon:
                    action = np.argmax(qtable[state,:])
                # Do the exploration
                else:
                    action = environment.action_space.sample()
                
                # Get the new state, reward, condition, and information
                new_state, reward, done, info = environment.step(action)

                # Bellman equation
                qtable[state, action] = qtable[state, action] + learning_rate * (reward + gamma *
                                                     np.max(qtable[new_state, :]) - qtable[state, action])
                
                # Add the reward to the total
                total_rewards += reward
                state = new_state
                
                # If we are done, return the new qtable
                if done == True:
                    break
            
            # Reduce the epsilon
            epsilon = min_epsilon + (max_epsilon - min_epsilon) * np.exp(-decay_rate * episode)
            # Append total rewards of the episode to the rewards
            rewards.append(total_rewards)
        
        # Return the qtable
        return qtable
```

```python
# Q-learn the agent
qtable = qlearn(environment, state_size, action_size, qtable, total_episodes, max_steps, learning_rate, gamma, epsilon,
                                                                 max_epsilon, min_epsilon, decay_rate)
```

```python
environment.reset()

for episode in range(5):
    state = environment.reset()
    step = 0
    done = False
    
    print("----------------------------------------------------")
    print("EPISODE ", episode)
    
    for step in range(max_steps):
        action = np.argmax(qtable[state,:])
        new_state, reward, done, info = environment.step(action)
        if done:
            environment.render()
            print("Number of steps", step)
            break
        state = new_state

environment.close()
```
